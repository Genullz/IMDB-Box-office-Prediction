```python
import pandas as pd
import numpy as np
import os, time,json
import seaborn as sns
```

```python
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2000.head()
```

```python
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
df_2001.head()
```

```python
df = pd.concat([df_2000,df_2001] )
df.head()
```

```python
finance_info = (df['budget']> 0) | (df['revenue']>0)
```

```python
finance_info.value_counts()
```

```python
eda = df[finance_info].copy()
```

```python
eda['certification'].value_counts()
```

```python
sns.countplot(data = eda, x = 'certification')
```

### What is the average revenue per certification category?

```python
eda.groupby('certification')['revenue'].mean()
```

```python
sns.barplot(data = eda, x = 'certification', y = 'revenue')
```

### What is the average budget per certification category?

```python
eda.groupby('certification')['budget'].mean()
```

```python
sns.barplot(data= eda, x = 'certification', y = 'budget')
```
